### Step 1. Importamos Pandas

Siempre que trabajamos con Pandas, lo primero es importarlo. Por convención se importa con el alias `pd` para no tener que escribir `pandas` cada vez.

````python
import pandas as pd
````

### Step 2. Datos en forma de diccionario

Los datos ya nos los dan. Un diccionario en Python es una estructura `{clave: valor}`. Aquí cada clave es el nombre de una columna, y cada valor es una lista con los datos de esa columna.

````python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
````

### Step 3. Crear un DataFrame y asignarlo a la variable `army`

`pd.DataFrame(raw_data)` convierte el diccionario en una tabla (DataFrame). Piénsalo así:
- El **diccionario** tiene claves → esas claves se convierten en los **nombres de las columnas**
- Cada **lista** de valores → se convierte en los **datos de esa columna**
- Pandas automáticamente le asigna un índice numérico (0, 1, 2...) a cada fila

````python
# pd.DataFrame() es la función que convierte cualquier estructura de datos compatible en una tabla
army = pd.DataFrame(raw_data)

# Mostramos las primeras filas para verificar que se creó bien
army
````

### Step 4. Poner la columna 'origin' como índice

Por defecto, el índice es numérico (0, 1, 2...). Con `set_index('origin')` le decimos a Pandas:
*"usa la columna origin como las etiquetas de las filas"*.

Esto es útil porque después podremos buscar filas por nombre de estado (ej: `army.loc['Texas']`) en lugar de por número.

**Importante:** `set_index()` no modifica el DataFrame original, devuelve uno nuevo. Por eso reasignamos con `army = army.set_index('origin')`.

````python
# set_index('origin') convierte la columna 'origin' en el índice del DataFrame
# Es decir, los nombres de los estados pasan a ser las "etiquetas" de las filas
army = army.set_index('origin') # Convertimos army = con la nueva distribución porque si no, con set_index() no modifica el DataFrame original.

army
````

### Step 5. Mostrar solo la columna 'veterans'

Para seleccionar **una sola columna** de un DataFrame usamos corchetes con el nombre entre comillas: `army['veterans']`

Esto devuelve una **Serie** (no un DataFrame). Una Serie es como una columna individual, con su índice (los nombres de los estados) y sus valores.

````python
# army['veterans'] -> selecciona la columna llamada 'veterans'
# El resultado es una Serie: una lista de valores con etiquetas (el índice)
army['veterans']
````

### Step 6. Mostrar las columnas 'veterans' y 'deaths'

Para seleccionar **múltiples columnas** pasamos una **lista** de nombres dentro de los corchetes.
Fíjate en el doble corchete: `army[['veterans', 'deaths']]`
- El corchete exterior → le dice a Pandas que vamos a seleccionar columnas
- El corchete interior → es la lista Python con los nombres de las columnas

El resultado ahora sí es un **DataFrame** (tiene más de una columna).

````python
# Doble corchete: army[[lista_de_columnas]]
# El interior es una lista Python ['col1', 'col2']
army[['veterans', 'deaths']]
````

### Step 7. Mostrar el nombre de todas las columnas

El atributo `.columns` de un DataFrame nos devuelve un objeto Index con los nombres de todas las columnas.
No es un método (no lleva paréntesis), es una propiedad del DataFrame.

````python
# .columns es un atributo (propiedad) del DataFrame, no un método
# Devuelve un objeto Index con todos los nombres de columnas
army.columns
````

### Step 8. Seleccionar columnas 'deaths', 'size' y 'deserters' de Maine y Alaska

Aquí necesitamos seleccionar filas **Y** columnas específicas a la vez. Para eso usamos **`.loc[]`**

`.loc[]` trabaja con **etiquetas** (nombres). Su sintaxis es:
```
army.loc[filas, columnas]
```
- `filas` → lista de etiquetas de fila: `['Maine', 'Alaska']`
- `columnas` → lista de nombres de columna: `['deaths', 'size', 'deserters']`

`.loc` es el selector cuando conoces los **nombres** (etiquetas).

````python
# .loc[filas, columnas] -> selección por ETIQUETAS (nombres)
# Primero le decimos qué filas queremos (por nombre de estado)
# Luego qué columnas queremos (por nombre de columna)
army.loc[['Maine', 'Alaska'], ['deaths', 'size', 'deserters']]
````

### Step 9. Seleccionar filas 3 a 7 y columnas 3 a 6

Aquí seleccionamos por **posición numérica** (número de fila/columna), no por nombre.
Para eso usamos **`.iloc[]`** (la "i" es de "integer").

`.iloc[]` usa la misma lógica de slicing que NumPy: `inicio:fin` donde **el fin no está incluido**.
- `iloc[3:7]` → filas en posición 3, 4, 5, 6 (no la 7)
- `iloc[:, 3:6]` → columnas en posición 3, 4, 5 (no la 6)

**Resumen clave:**
- `.loc[]` → por **nombres/etiquetas**
- `.iloc[]` → por **posición numérica** (0, 1, 2...)

````python
# .iloc[filas, columnas] -> selección por POSICIÓN (números enteros)
# 3:7 significa posiciones 3, 4, 5, 6 (el 7 no se incluye, igual que en NumPy)
army.iloc[3:7, 3:6]
````

### Step 10. Seleccionar todas las filas después de la cuarta y todas las columnas

Con `.iloc` podemos hacer slicing:
- `4:` → desde la posición 4 hasta el final (sin límite superior)
- `:` → todas las columnas (sin límite inferior ni superior)

Es exactamente igual que el slicing de listas en Python: `lista[4:]`

````python
# 4: -> desde la fila en posición 4 hasta el final
# : -> todas las columnas (sin restricción)
army.iloc[4:, :]
````

### Step 11. Seleccionar hasta la 4ª fila (incluida) y todas las columnas

- `:4` → desde el inicio hasta la posición 4 (la posición 4 NO se incluye, así que obtenemos 0, 1, 2, 3)
- `:` → todas las columnas

````python
# :4 -> desde el inicio hasta la posición 4 (sin incluir la 4, igual que slicing Python). Si fuese 4 incluido sería :5
# Obtenemos las filas 0, 1, 2, 3 → las primeras 4 filas
army.iloc[:4, :]
````

### Step 12. Seleccionar desde la columna 3 hasta la 7

Seleccionamos todas las filas (`:`) y las columnas desde la posición 3 hasta la 6 (recordar: 7 no se incluye).

````python
# : -> todas las filas
# 3:7 -> columnas en posición 3, 4, 5, 6
army.iloc[:, 3:7]
````

### Step 13. Seleccionar filas donde deaths > 50

Esto es **filtrado booleano** (o "masking"), igual que en NumPy.
Funciona en 2 pasos:
1. `army['deaths'] > 50` → crea una Serie de True/False para cada fila
2. `army[...]` → pasa esa Serie como índice, y Pandas solo devuelve las filas donde es True

Es como decirle a Pandas: *"dame solo las filas donde esta condición es verdadera"*

````python
# Paso 1: army['deaths'] > 50 genera una máscara de True/False
# Paso 2: army[máscara] filtra y devuelve solo las filas donde la condición es True
army[army['deaths'] > 50]
````

### Step 14. Seleccionar filas donde deaths > 500 OR deaths < 50

Para combinar múltiples condiciones en Pandas:
- **OR** → se escribe con `|` (tubería/pipe)
- **AND** → se escribe con `&`

**Importante:** Cada condición individual DEBE ir entre paréntesis cuando las combinamos. Sin los paréntesis, Python se confunde con el orden de operaciones.

````python
# | es el operador OR en Pandas (no se usa 'or')
# Cada condición va entre paréntesis obligatoriamente
army[(army['deaths'] > 500) | (army['deaths'] < 50)]
````

### Step 15. Seleccionar todos los regimientos que NO sean "Dragoons"

Para negar una condición usamos `~` (tilde), que es el operador NOT en Pandas.
También podemos usar `!=` (diferente de) para una comparación directa de desigualdad.

````python
# Opción 1: != (diferente de) -> directamente compara que no sea igual a 'Dragoons'
army[army['regiment'] != 'Dragoons']
````

````python
# Opción 2: ~ (tilde? virgulilla?) niega la condición completa
# ~ es el NOT booleano en Pandas (equivale a 'no es verdad que...')
army[~(army['regiment'] == 'Dragoons')]
````

### Step 16. Seleccionar las filas llamadas Texas y Arizona

Como el índice del DataFrame son los nombres de los estados (lo cambiamos en el Step 4), podemos seleccionar filas por nombre usando `.loc[]`.

Pasamos una lista con los nombres que queremos.

````python
# Como el índice del DataFrame son los nombres de los estados (lo cambiamos en el Step 4), podemos seleccionar filas por nombre usando `.loc[]`.
# .loc con una lista de etiquetas -> devuelve las filas con esos nombres de índice
army.loc[['Texas', 'Arizona']]
````

### Step 17. Seleccionar la tercera celda en la fila de Arizona

Necesitamos una fila específica (por nombre) y dentro de ella, una celda específica (por posición).

Podemos combinar `.loc` y `.iloc`:
1. `army.loc['Arizona']` → selecciona la fila de Arizona (devuelve una Serie)
2. `.iloc[2]` → de esa Serie, toma el elemento en posición 2 (la tercera, porque empieza en 0)

**Recuerda:** en Python los índices empiezan en 0: posición 0 = primera, posición 2 = tercera.

````python
# Paso 1: army.loc['Arizona'] -> selecciona la fila Arizona como una Serie
# Paso 2: .iloc[2] -> de esa Serie, toma el elemento en posición 2 (el tercero)
# Posición 0=regiment, 1=company, 2=deaths -> el tercero es 'deaths'
army.loc['Arizona'].iloc[2]
````

### Step 18. Seleccionar la tercera celda hacia abajo en la columna deaths

Ahora al revés: seleccionamos primero la columna (por nombre) y luego una fila (por posición).

1. `army['deaths']` → selecciona la columna deaths (devuelve una Serie)
2. `.iloc[2]` → de esa Serie, toma el elemento en posición 2 (la tercera fila)

````python
# Paso 1: army['deaths'] -> selecciona la columna deaths como una Serie
# Paso 2: .iloc[2] -> de esa columna, toma el valor en la tercera posición (índice 2)
# Eso corresponde a Texas (fila 0=Arizona, 1=California, 2=Texas)
army['deaths'].iloc[2]
````

## 📋 Resumen de conceptos clave del ejercicio

| Operación | Sintaxis | Cuándo usarla |
|-----------|----------|---------------|
| Crear DataFrame | `pd.DataFrame(datos)` | Siempre que quieras crear una tabla |
| Cambiar índice | `df.set_index('columna')` | Cuando una columna tiene mejores etiquetas que números |
| Una columna | `df['columna']` | Para seleccionar una columna → devuelve Serie |
| Varias columnas | `df[['col1', 'col2']]` | Para seleccionar varias → devuelve DataFrame |
| Nombres de columnas | `df.columns` | Para ver qué columnas existen |
| Selección por etiqueta | `df.loc[filas, columnas]` | Cuando conoces los **nombres** |
| Selección por posición | `df.iloc[filas, columnas]` | Cuando conoces los **números** |
| Filtrado simple | `df[df['col'] > valor]` | Para filtrar por una condición |
| Filtrado OR | `df[(cond1) \| (cond2)]` | Para filtrar con dos condiciones alternativas |
| Filtrado NOT | `df[~(condicion)]` o `df[df['col'] != valor]` | Para negar una condición |
