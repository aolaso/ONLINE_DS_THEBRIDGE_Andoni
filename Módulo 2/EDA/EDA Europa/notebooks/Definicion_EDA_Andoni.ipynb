## 1. Descripción

Este proyecto analiza la evolución del **salario mínimo en los países de la Unión Europea** desde una perspectiva social y económica: qué puedes comprar realmente con él, cuánto te queda después de pagar el alquiler, cómo afecta a la desigualdad de género, y si tener trabajo es suficiente para no ser pobre.

La pregunta central es directa y muy actual: **¿sirve de algo el salario mínimo en Europa hoy?**

El hilo conductor del proyecto es una paradoja que los datos europeos evidencian con claridad:

> *El salario mínimo ha subido en casi todos los países de la UE durante la última década. Sin embargo, el poder adquisitivo real de los trabajadores con menor renta no ha mejorado al mismo ritmo — porque el coste de vida, especialmente la vivienda, ha crecido mucho más rápido que los salarios.*

Subir el número no es suficiente si todo lo demás sube más. Eso es lo que este análisis intenta demostrar con datos.

---
## 2. Hipótesis

Las hipótesis se plantean **antes de ver los datos** — eso es fundamental en un EDA riguroso. Al final del proyecto comprobaremos cuáles se confirman y cuáles no.

---

**H1 — El salario mínimo nominal engaña** *(hipótesis principal)*  
Los países con mayor salario mínimo en euros no son necesariamente los que ofrecen mayor poder adquisitivo real. Al ajustar por el coste de vida mediante PPS (Purchasing Power Standard), las diferencias entre países se reducen drásticamente.

---

**H2 — El salario mínimo es una herramienta de política feminista**  
Los países con un salario mínimo más alto como porcentaje del salario medio tienen una brecha salarial de género menor. El sistema ha colocado históricamente a las mujeres en los empleos peor pagados. Subir el suelo salarial corrige estructuralmente esa injusticia.

---

**H3 — En la mayoría de la UE, cobrar el mínimo significa no poder pagar el alquiler**  
El salario mínimo no es suficiente para vivir de forma independiente: en la mayoría de países europeos implica destinar más del 40% del sueldo al alquiler, haciendo imposible la emancipación para millones de jóvenes.

---

**H4 — Working poor: tener trabajo no te salva de la pobreza**  
Los países con un salario mínimo más bajo en términos reales concentran mayor proporción de trabajadores en riesgo de exclusión social. El empleo sin salario digno no cumple su función protectora.

---

**H5 — La Generación Z es la primera que trabaja más para vivir peor que sus padres**  
Los salarios mínimos han subido en casi todos los países de la UE, pero la combinación de alquileres disparados y estancamiento salarial real hace que la generación Z tenga peor capacidad adquisitiva que sus padres a la misma edad.

## 5. Plan de análisis y justificación del merge

Cada dataset por separado tiene entre 38 y 330 filas. Para alcanzar el mínimo de **1.000 filas y 10 columnas** necesario para un análisis exploratorio robusto, se realizará un **merge** de los 5 datasets en uno único.

Un merge es la operación de combinar dos o más tablas usando una columna común como clave. En este caso la clave será el **país** y el **año** — cada fila del DataFrame final representará un país en un año concreto, con todas sus variables asociadas.

El DataFrame final tendrá esta estructura aproximada:

| pais | anio | salario_minimo_eur | pib_pps | indice_vivienda | gini | brecha_genero |
|---|---|---|---|---|---|---|
| Spain | 2020 | 950 | 91 | 112.3 | 33.0 | 11.9 |
| Germany | 2020 | 1584 | 118 | 108.7 | 31.5 | 18.3 |
| ... | ... | ... | ... | ... | ... | ... |

El EDA completo seguirá esta estructura:

1. **Limpieza de datos** — formato de columnas, valores nulos, tipos de datos
2. **Merge de los 5 datasets** — construcción del DataFrame unificado
3. **Análisis univariante** — distribución de cada variable por separado
4. **Análisis bivariante** — relaciones entre pares de variables
5. **Análisis multivariante** — correlaciones, heatmaps, pairplots
6. **Contraste de hipótesis** — ¿se confirman las 5 hipótesis?
7. **Conclusiones (y plan de acción)**
