17. Drop rows with *Nan* in `titanic` with `how='any'` and print the shape



```python
# dropna() elimina filas que contienen valores NaN (datos que faltan)
# how='any' significa: elimina la fila si CUALQUIER columna tiene NaN
# Es decir, basta con que falte un solo dato para que la fila desaparezca
titanic_any = titanic.dropna(how='any')

print(f"Shape original: {titanic.shape}")
print(f"Shape con how='any': {titanic_any.shape}")

# shape[0] coge solo el número de filas de la tupla (filas, columnas)
print(f"Filas eliminadas: {titanic.shape[0] - titanic_any.shape[0]}")
```

19. Drop columns in `titanic` with more than 1000 missing values and print the columns remaining

```python
# isnull() genera un DataFrame de True/False donde True = NaN
# .sum() suma los True de cada columna (True=1, False=0)
# El resultado es el número de NaN por columna
nulos_por_columna = titanic.isnull().sum()
print("NaN por columna:")
print(nulos_por_columna)
print("")

# axis=1 significa que eliminamos COLUMNAS (no filas)
# thresh = número mínimo de valores NO nulos para que la columna se quede
# ESTO CREO QUE ESTÁ MAL: len(titanic) - 1000 significa: necesita al menos (total filas - 1000) valores válidos
# CREO QUE MAL TAMBIÉN: Equivale a eliminar columnas con más de 1000 NaN 
# # thresh=len(titanic) - 1000 significa:
# la columna se queda si tiene AL MENOS (total_filas - 1000) valores válidos
# NO es exactamente lo mismo que "eliminar columnas con más de 1000 NaN"
# porque depende del total de filas del dataset
titanic_limpio = titanic.dropna(axis=1, thresh=len(titanic) - 1000)

print("Columnas restantes después de eliminar las que tienen más de 1000 NaN:")
print(titanic_limpio.columns.tolist())
```

20. Calculate the ratio of missing values at the `boat` column. 

```python
# isnull() marca con True cada NaN de la columna
# .sum() cuenta los True (los NaN)
# Dividimos entre el total de filas para obtener la proporción
ratio_nulos = titanic['boat'].isnull().sum() / len(titanic)

print(f"Valores NaN en 'boat': {titanic['boat'].isnull().sum()}")
print(f"Total de filas: {len(titanic)}")

# :.2% convierte automáticamente el decimal a porcentaje con 2 decimales
# 0.623 → 62.30%
print(f"Ratio de NaN en 'boat': {ratio_nulos:.2%}")
```

21.  Group `titanic` by `'pclass'` and aggregate by the columns `age` & `fare`, by `max` and `median` and assign it to `by_class`


```python
# groupby('pclass') agrupa por clase (1 = primera, 2 = segunda, 3 = tercera)
# [['age', 'fare']] selecciona dos columnas a la vez (doble corchete = lista)
# .agg(['max', 'median']) aplica dos funciones a la vez sobre esas dos columnas
# El resultado tiene columnas con dos niveles: (age/fare) x (max/median)
by_class = titanic.groupby('pclass')[['age', 'fare']].agg(['max', 'median'])

print("Estadísticos de edad y precio de billete por clase:")
print(by_class)
```

24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
# pivot_table crea una tabla resumen con dos variables cruzadas
# values='survived': qué columna queremos resumir (0=murió, 1=sobrevivió)
# index='pclass': qué columna se convierte en las filas de la tabla
# columns='gender': qué columna se convierte en las columnas de la tabla
# aggfunc='sum': sumamos los 1s de survived → nos da el número de supervivientes
counted = titanic.pivot_table(
    values='survived',
    index='pclass',
    columns='gender',
    aggfunc='sum'
)

print("Supervivientes por clase y género:")
counted
```

27. Please, show only the rows using a mask with the following conditions: 
    - They are woman
    - From third class
    - Younger than 30
    - They survived
    
    ¿How many rows fulfill the condition?

```python
# Una máscara es un filtrado booleano con varias condiciones combinadas
# Cada condición va entre paréntesis (obligatorio cuando combinas con &)
# & significa "Y": las dos condiciones tienen que ser True a la vez
mask = (
    (titanic['gender'] == 'female') &   # que sea mujer
    (titanic['pclass'] == 3) &           # de tercera clase
    (titanic['age'] < 30) &              # menor de 30 años
    (titanic['survived'] == 1)           # que sobrevivió
)

resultado = titanic[mask]

print(f"Mujeres de 3ª clase, menores de 30 y supervivientes: {len(resultado)}")
print(resultado[['name', 'age', 'pclass', 'gender', 'survived']])
```

32. Using `matplotlib` find the appropriate plot to visualize the column `'gender'`

```python
# 'gender' es una variable categórica: solo tiene dos valores (male/female)
# Para variables categóricas el gráfico apropiado es el de barras (bar chart)
# Un histograma no tiene sentido aquí porque no hay rangos numéricos

# Primero contamos cuántos hay de cada género
conteo_genero = titanic['gender'].value_counts()

# .plot(kind='bar') convierte la Serie en un gráfico de barras
conteo_genero.plot(kind='bar', color=['steelblue', 'salmon'], edgecolor='black')

plt.xlabel('Género')
plt.ylabel('Número de pasajeros')
plt.title('Distribución de género en el Titanic')
plt.xticks(rotation=0)  # rotation=0 deja las etiquetas horizontales, más fáciles de leer
plt.show()
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
# Primero filtramos los que no sobrevivieron (survived == 0)
# Luego cogemos solo la columna age
# dropna() elimina los NaN de age porque kdeplot no puede trabajar con huecos
no_sobrevivientes = titanic[titanic['survived'] == 0]['age'].dropna()

# kdeplot dibuja una curva suave que representa la distribución de los datos
# Es como un histograma pero en vez de barras dibuja una línea continua
# fill=True rellena el área bajo la curva con color
# (el enunciado dice shade=True pero en versiones modernas se llama fill=True)
sns.kdeplot(
    no_sobrevivientes,
    color='red',
    label='Not Survived',
    fill=True
)

plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.title('Distribución de edad de los no supervivientes')
plt.legend()  # muestra la leyenda con el label 'Not Survived'
plt.show()
```
