## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

def devuelve_duplicados(lista, duplicado):
    indices = []
    if lista.count(duplicado) > 1:
        for i in range(len(lista)):
            if lista[i] == duplicado:
                indices.append(i)
    return indices

print(devuelve_duplicados(ejer_3, "un"))
print(devuelve_duplicados(ejer_3, "es"))
print(devuelve_duplicados(ejer_3, "binario"))
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def comprobar_dni(numero, letra_usuario):
    # 1. Convertimos el número a "texto" para poder contar cuántos dígitos tiene
    numero_a_texto = str(numero)
    largo = len(numero_a_texto)

    # 2. Comprobamos si el largo está entre 7 y 8 (regla del ejercicio)
    # Si no cumple esto, devolvemos False directamente y la función termina
    if largo < 7 or largo > 8:
        return False

    # 3. Esta es la lista de letras oficial en el orden correcto
    palabra = 'TRWAGMYFPDXBNJZSQVHLCKE'

    # 4. Calculamos la posición (el resto de dividir el número entre 23)
    # El símbolo % es el "módulo": nos da lo que sobra de la división
    posicion = numero % 23

    # 5. Sacamos la letra que le corresponde a esa posición
    letra_correcta = palabra[posicion]

    # 6. Comparamos la letra que nos han dado con la que hemos calculado
    # Usamos .upper() por si el usuario escribió la letra en minúscula
    if letra_usuario.upper() == letra_correcta:
        return True
    else:
        return False

# --- EJEMPLOS PARA PROBAR EL CÓDIGO ---
# (Puedes borrar esto si solo quieres la función)
print(comprobar_dni(55555555, 'Z'))  # Debería dar True
print(comprobar_dni(123, 'A'))       # Debería dar False (es muy corto)
`````
