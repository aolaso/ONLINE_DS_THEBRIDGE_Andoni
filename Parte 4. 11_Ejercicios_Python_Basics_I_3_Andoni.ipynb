## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1

print(A == B)
print(A != C)
print(A > C)
print(C <= A)
print(B != C)
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input_1 = input("Di algo")
input_2 = input("Di algo más (puedes repetir lo que has dicho previamente o algo nuevo)")
input_3 = input("Di una última cosa (también igual o diferente a las anteriores)")

if input_1 == input_2 == input_3:
    print("Todas las respuestas son iguales")
else:
    print("No todas las respuestas son iguales")

print("*" * 30)

if (input_1 == input_2) or (input_1 == input_3) or (input_2 == input_3):
    print("Al menos dos de las respuestas son iguales")
else:
    print("No hay dos respuestas iguales")
````

## Ejercicio 4
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

1. Tenemos True and True, que da True, y también True and False, que da False. Como se está usando and, en cuanto aparece un solo False, el resultado completo pasa a ser False.

2. Si analizamos True or False, el resultado es True, porque con or basta con que uno de los dos valores sea True. Pero al poner not delante, ese valor se invierte y el resultado final se convierte en False.

3. Dentro del paréntesis hay un or que incluye un True, así que esa parte vale True. Sin embargo, fuera aparece un and False, y eso hace que toda la expresión termine siendo False.

4. Aquí, not False se transforma en True, y junto con and True sigue dando True. Pero luego aparece not True, que es False, y como todo está conectado con and, ese único False hace que el resultado final sea False.
