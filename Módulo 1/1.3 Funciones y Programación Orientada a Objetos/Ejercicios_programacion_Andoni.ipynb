### 2. Filtrar números pares de una lista:

Emplea filter para obtener sólo los números pares de la lista [1, 2, 3, 4, 5, 6].

Solución esperada: [2, 4, 6]

```python
lista2 = [1,2,3,4,5,6]

lista_pares = filter(lambda x: x % 2 == 0, lista2)

print(list(lista_pares))
```

### 3. Convertir grados Celsius a Fahrenheit:

Usa map y una función lambda para convertir una lista de temperaturas en grados Celsius [0, 10, 20, 30] a Fahrenheit.

Solución esperada: [32.0, 50.0, 68.0, 86.0]

```python
celsius = [0,10,20,30]

fahrenheit = map(lambda x: x*1.8 + 32, celsius)

print(list(fahrenheit))
```

### 4. Sumar los cuadrados de los números pares de una lista:

Combina filter, map y reduce (importa reduce desde functools) para sumar los cuadrados de los números pares en [1, 2, 3, 4, 5, 6].

Solución esperada: 56

```python
from functools import reduce
```

```python
lista1 = [1,2,3,4,5,6]

# 2. Aplicamos filter para quedarnos con los pares (donde el resto de dividir por 2 es 0)
pares = filter(lambda x: x % 2 == 0, lista1)


# 3. Aplicamos map para elevar esos pares al cuadrado
cuadrados = map(lambda x: x**2, pares)

# 4. Aplicamos reduce para sumar los elementos de la lista de cuadrados
# Reduce coge el primero (4), le suma el siguiente (16), y al resultado le suma el último (36)
resultado = reduce(lambda x,y: x+y, cuadrados)

print(f"El resultado final es: {resultado}")
```

```python
# Vamos a intentar hacerlo todo en una sola línea (con ayuda por supuesto):
from functools import reduce

lista1 = [1, 2, 3, 4, 5, 6]

resultado = reduce(lambda x, y: x + y, map(lambda x: x**2, filter(lambda x: x % 2 == 0, lista1)))

print(f"El resultado final es: {resultado}")
```

### 5. Encadenar map, filter, y reduce para resolver un problema complejo:

Dada una lista de strings que representan números enteros y "NA" (por ejemplo, ["2", "5", "NA", "9", "10"]), convierte los números a enteros, filtra los "NA", suma el cuadrado de los números que son mayores que 5 y menores que 10.

Solución esperada: Un único valor que es la suma de los cuadrados de los números válidos.

```python
lista_original = ["2", "5", "NA", "9", "10"]
from functools import reduce
```

```python
# 1. Filtramos los "NA", para que nos devulva una lista sin "NA". Usamos 'filter'
lista_sin_NA = list(filter(lambda x: x != "NA", lista_original))
print(lista_sin_NA)
```

```python
#. Convertimos los strings (STR) de los números de la lista original en números enteros (INT). Usamos 'map'
lista_num = list(map(lambda x: int(x), lista_sin_NA))
print(lista_num)
```

```python
# ¡¡¡¡MAL!!!! (lo dejo para aprender)
resultado = reduce(lambda x,y: ((x < 10 and > 5 )**2), lista_num) # He intentado hacer todo el una sola línea pero NO se puede, 
                                                                  # hay que filtrar primero los números entre 5 y 10 con 'filter'
```

```python
# Filtramos los números entre 5 y 10 usando 'filter'. Esto como paso intermedio antes de hacer el cuadrado y después usar 'reduce'
num_selec = list(filter(lambda x: 5 < x < 10, lista_num)) #IMPORTANTE: meter 'list' aquí y NO cuando haces el 'print' porque si no se "vacía el cartucho". 
print(num_selec)
```

```python
# Segundo paso intermedio (aunque se podría usar ya 'reduce' y hacerlo todo en una sola línea, pero me parece demasiado complicado)
num_cuad = list(map(lambda x: x**2, num_selec))
print(num_cuad)
```

```python
# Usamos 'reduce', aunque como solo hay un elemento, no se va a sumar a nada: 
resultado = reduce(lambda x,y: x+y, num_cuad)
print(resultado)
```

```python
# En una sola línea con 'reduce' + haciendo el cuadrado de los números: 
resultado = reduce(lambda x,y: x+y, map(lambda x: x**2, filter(lambda x: 5 < x < 10, lista_num))) # aquí hemos metido en el x y el y de lambda (de resultado) 
                                                                                                  # las lambdas de 'num_selec' y 'num_cuad'
print(resultado)
```
