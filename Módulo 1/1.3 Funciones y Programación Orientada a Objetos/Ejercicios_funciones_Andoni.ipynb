## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def dia_de_la_semana(numero):
    dias = {1: "Lunes", 2: "Martes", 3: "Miércoles", 4: "Jueves", 5: "Viernes", 6: "Sábado", 7: "Domingo"}
    
    # Preguntamos: ¿Está el número en las llaves del diccionario?
    if numero in dias:
        return dias[numero] # Si está, nos devuelve el valor
    else:
        return "Número inválido (debe ser del 1-7)" # Si no está, nos devuelve el error
    
````

````python
print(dia_de_la_semana(7))
print(dia_de_la_semana(10))
print(dia_de_la_semana(3))
````

## Ejercicio 3
Escribe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero sea  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def comparar_numeros(num1, num2):
    # Caso 1: Son iguales
    if num1 == num2:
        return "Son iguales"
    
    # Caso 2: El primero gana
    elif num1 > num2:
        return "El primero es mayor que el segundo"
    
    # Caso 3: Solo queda que el segundo gane
    else:
        return "El segundo es mayor que el primero"

print(comparar_numeros(9, 34))
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea una letra que contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayúscula, como minúscula

````python
def contador_letras(texto, letra_target):
    cuenta = 0
    # Ponemos todo en minúsculas para no fallar con las mayúsculas
    texto = texto.lower()
    letra_target = letra_target.lower()
    
    # Vamos letra por letra cotejando
    for letra_actual in texto:
        if letra_actual == letra_target:
            cuenta = cuenta + 1 # Si es la letra, sumamos 1 al marcador
            
    return cuenta


resultado = contador_letras("Me llamo Andoni y me parece que aprender Python es muy complicado", "o")

print(f"Las veces que se ha encontrado la letra 'O' es: {resultado}")
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string, siendo la clave la letra y el valor el conteo.

````python
def dic_conteo (texto):
    # 1. Creamos donde guardar el resultado.
    resultado = {}
    
    # 2. Empiezo a recorrer el texto letra por letra.
    for letra in texto:
        
        # 3. Pregunto: ¿Esta letra NO está anotada todavía?
        if letra not in resultado:
            # Como es la primera vez que la veo, la anoto con un 1.
            # Es como decir: "He visto una 'p' por ahora".
            resultado[letra] = 1
            
        # 4. Si la pregunta de arriba es mentira (significa que SÍ está):
        else:
            # Busco el número que ya tenía anotado y le sumo 1.
            # Si ya tenía 'a': 1, ahora pasará a ser 'a': 2.
            resultado[letra] = resultado[letra] + 1
            
    # 5. Cuando el bucle termina de recorrer todo el texto, 
    # devuelvo la libreta rellena.
    return resultado

# 6. Probamos la función y guardamos el resultado
frase = "esternocleidomastoideo"
mi_conteo = dic_conteo(frase)

# 7. Lo mostramos por pantalla
print(mi_conteo)
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

Tendrá como salida la lista

````python
def gestionar_lista(lista, comando, elemento=None):
    # PASO A: ¿La orden es añadir?
    if comando == "add":
        lista.append(elemento) # .append es "pegar al final"
        
    # PASO B: ¿La orden es quitar?
    elif comando == "remove":
        # Antes de borrar, el miramos si el objeto está en la caja
        if elemento in lista:
            lista.remove(elemento) # .remove es "sacar de la caja"
        else:
            # Si no está, te avisa para que no haya errores
            print(f"¡Error! No puedo quitar '{elemento}' porque no está en la lista.")
    
    # PASO C: Devolvemos la caja modificada
    return lista
````

## Ejercicio 8
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

````python
def fibonacci(n):
    # El límite por abajo: Si el número es 0 o 1, devolvemos el mismo número, ya que no puede seguir bajando. 
    # Sin esto, el programa se quedaría contando números negativos para siempre, bucle infinito (creo?)
    if n <= 1:
        return n
    else:
        # la regla de Fibonacci: Suma el resultado de (n-1) y (n-2)
        return fibonacci(n - 1) + fibonacci(n - 2)


print(f"El número en la posición 9 es: {fibonacci(9)}")
````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el área de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math #definimos funciones matemáticas 

def area_cuadrado(lado):
    # Área cuadrado: Un cuadrado es lado por lado
    return lado ** 2

def area_triangulo(base, altura):
    # Área triángulo: La mitad de un rectángulo (base * altura / 2)
    return (base * altura) / 2

def area_circulo(radio):
    # Área círculo: Pi por el radio al cuadrado
    return math.pi * (radio ** 2)
````

````python
resultado_1 = (2 * area_circulo(10)) + area_triangulo(3,7)

resultado_2 = (area_cuadrado(10) + area_circulo(4) + (2 * area_circulo(6)) + (5 * area_triangulo(2,4)))  

print(resultado_1)
print(resultado_2)
````
