## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda: 
    '''
    Clase para gestionar nuestras tiendas de electrodomésticos.
    '''

    tipo = "Electrodomesticos"
    abierta = True
    
    '''
    Definimos el constructor específico para cada tienda 
    '''
    def __init__(self, nombre, direccion, num_empleados, ventas_3_meses):
        self.nombre = nombre # String
        self.direccion = direccion #String
        self.num_empleados = num_empleados #int
        self.ventas = ventas_3_meses # Lista de números # Creo que se puede poner otro nombre más corto como 'Ventas' aquí después de self. (preguntar en clase)
    
    def total_ventas (self):
        '''
        Devuelve la suma de ventas de los 3 meses
        '''
        total= sum(self.ventas)
        return total
 
    def media_ventas_empleado (self):  
        '''
        Devuelve media mensual de las ventas entre número de empleados
        '''
        total = sum(self.ventas)
        media = total / self.num_empleados
        return media
    
    def identidad_tienda (self):
        '''
        Nos devuelve un texto juntando el nombre y la dirección
        '''
        return self.nombre + " en " + self.direccion
    
    def ventas_ultimo_mes (self):
        '''
        Busca el último número de la lista de ventas
        '''
        return self.ventas[-1] # Para que nos devuelva el último elemento de la lista 
        
    def marketing (self, x): # Porque aquí tenemos que introducir un parámetro de entrada (que es el dinero que invertimos en márketin, en este caso)
        '''
        Si invertimos dinero (X), calculamos cuánto subirían las ventas
        '''
        
        '''
        Recalculamos ventas según inversión en marketing:
        x < 1000  → ventas * 1.2
        x >= 1000 → ventas * 1.5
        '''
        if x < 1000:
            multiplicador = 1.2
        else:
            multiplicador = 1.5

        # usamos comprensión: recorre la lista y multiplica cada venta (esto lo he entendido gracias a la IA, porque no sabía cómo hacerlo)
        '''
        Si tengo una tienda con ventas de [100, 200, 300] y llamo a la función con 500 euros (x = 500):

        Como 500 es menor que 1000, el factor es 1.2.

        La máquina hace: 100*1.2, 200*1.2 y 300*1.2.

        La lista de la tienda ahora pasa a ser: [120, 240, 360].
        '''
        self.ventas = [venta * multiplicador for venta in self.ventas]
        return self.ventas
        
```

```python
'''
Creamos las 3 tiendas con sus respectivos atributos: (Nombre, Dirección, Número de empleados, Ventas de los últimos 3 meses)
'''
tienda1 = Tienda("ElectroMad", "Avenida de la Cordura, 46", 10, [15000, 18000, 21000])
tienda2 = Tienda("TecnoBilbao", "Calle Gran Vía, 89", 7, [9000, 11000, 13000])
tienda3 = Tienda("MegaElectro", "Avenida del Atlaántico, 3", 5, [7000, 8500, 9200])

# Comprobaos todo en tienda1:
print("*** Comprobando tienda1 ***")
print("Tipo:", tienda1.tipo)
print("Abierta:", tienda1.abierta)
print("Nombre:", tienda1.nombre)
print("Dirección:", tienda1.direccion)
print("Empleados:", tienda1.num_empleados)
print("Ventas:", tienda1.ventas) 
print("Total ventas:", tienda1.total_ventas())
print("Media por empleado:", tienda1.media_ventas_empleado()) # ponemos() después del .-- porque aquí es donde se ejecutan los valores que hemos introducido en cada tienda
print("Nombre y dirección:", tienda1.identidad_tienda())
print("Ventas último mes:", tienda1.ventas_ultimo_mes())
print("Proyección con 500€ en marketing:", tienda1.marketing(500))

# Ventas del último mes de TODAS las tiendas con for:
print("\n*** Ventas último mes por tienda ***")
tiendas = [tienda1, tienda2, tienda3]
for tienda in tiendas:
    print(tienda.nombre, ">>", tienda.ventas_ultimo_mes())

# Tiendas cuya dirección lleve "Avenida":
print("\n*** Tiendas en Avenida **")
for tienda in tiendas:
    if "Avenida" in tienda.direccion:
        print(tienda.nombre)
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro: 
    """
    Clase que representa a un perro.
    
    Atributos comunes (todos los perros):
        patas: 4
        orejas: 2
        ojos: 2
        velocidad: 0 (el perro está parado por defecto)
    
    Parámetros del constructor _init_:
        raza (str): raza del perro
        pelo (str): color del pelo, por defecto "Marrón"
        dueño (str): nombre del dueño, por defecto None (sin dueño) - # pondremos dueno porque la Ñ es mejor no usarla 
    """
    # Atributos comunes: 
    patas = 4
    orejas = 2 
    ojos = 2
    velocidad = 0 
    
    def __init__(self, raza, color = "marrón", dueno = None): #IMPORTANTE METER 'MARRON' COMO STRING, PORQUE ME HA DADO ERROR HASTA QUE LO HE CAMBIADO. DESPISTE
        '''
        CONSTRUCTOR DEL PERRO:
        La raza hay que decirla siempre. El pelo y el dueño son opcionales. 
        '''
        self.raza = raza
        self.color = color 
        self.dueno = dueno 
        
    def andar (self, aumento_velocidad): # Para introducir un parámetro de entrada de la velocidad 
        '''
        FUNCIÓN ANDAR:
        Le suma velocidad a la que ya tenía el perro (que era 0)
        '''
        self.velocidad = self.velocidad + aumento_velocidad # Sumamos el aumento introducido 
        print("El perro ahora va a", self.velocidad, "km/por hora")
    
    def parar (self):
        '''
        FUNCIÓN PARAR: 
        Pone la velocidad del perro a 0 de nuevo
        '''
        self.velocidad = 0 # Se pone a 0 la velocidad
        print("El perro ahora va a", self.velocidad, "km/por hora, es decir, se ha parado")
        
    def ladrar (self, perro_habla):
        '''
        FUNCIÓN LADRAR:
        Hace que el perro diga ¡GUAU! seguido de lo que tú quieras. 
        '''
        return "¡GUAU!" + str(perro_habla)
        
```

```python
'''
Creamos nuestro perro (sin dueño)
'''
mi_perro = Perro("Corgi")
raza = mi_perro.raza
color = mi_perro.color
dueño = mi_perro.dueno

'''
Comprobamos los datos del perro 
'''
print(f"Mi perro es de la raza {raza}, es de color {color} y su dueño es {dueño} (es decir, no tiene dueño)" + "\n") 
# Quizás si lo hago con PRINT por cada atributo, queda tan raro como 'su dueño es None'

print("*" * 30 + "\n")

'''
Ahora, lo vamos a hacer andar metiendo un valor en aumento_velocidad, y después lo vamos a parar
'''
mi_perro.andar(40)
mi_perro.parar()

print("\n")
'''
Ahora vamos a hacer que nuestro perro hable, metiendo lo que queremos que diga dentro del argumento perro_habla
'''
print(mi_perro.ladrar("¡Quiero que me encuentres un dueño!"))
```
