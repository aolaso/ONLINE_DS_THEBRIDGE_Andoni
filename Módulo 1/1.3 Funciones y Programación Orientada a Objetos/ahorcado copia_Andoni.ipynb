## Juego del Ahorcado
El funcionamiento es que vaya mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _```. 

Si falla, se mostrará el número de errores que lleva y, si lleva 6 errores, entonces se mostrará "Fin del juego".
Encapsúlalo en una función.

````python
# No hubiera sido capaz de hacer esto solo sin la ayuda de la IA o de mis compañeros.

def ahorcado():
    # 1. Variables cortas y claras
    errores = 0
    palabra = "TIGRE"
    let_acertadas = set() # La bolsa de aciertos

    print("¡Bienvenidx al Juego del Ahorcado! (X_X)")

    while errores < 6:
        
        # FABRICAMOS EL DIBUJO PASO A PASO 
        dibujo = "" # Empezamos con el dibujo vacío
        
        for i in palabra: # Miramos cada letra de la palabra. La letra es = i
            if i in let_acertadas: # ¿La hemos acertado?
                dibujo = dibujo + i + " " # Ponemos la letra
            else:
                dibujo = dibujo + "_ " # Ponemos la raya
        
        print(f"\nPalabra: {dibujo}")
        print(f"Errores: {errores}/6")

        # COMPROBAR SI HEMOS GANADO 
        if "_" not in dibujo: # Si ya no hay rayas, se gana el juego
            print(f"¡HAS GANADO! 👏🏻 La palabra era: {palabra}")
            return 

        # PEDIR LETRA 
        letra = input("Introduce una letra: ").upper() # Convertimos todas en mayúsculas para que no haya errores 

        # LÓGICA DE LA LETRA 
        if letra in let_acertadas:
            print("Esta letra ya la habías puesto.")
            
        elif letra in palabra:
            let_acertadas.add(letra) # Guardamos el acierto en la bolsa (set)
            print(f"¡Bien! La '{letra}' está en la palabra que buscas.")
            
        else:
            errores = errores + 1 # Sumamos un fallo
            print(f"La '{letra}' no está. Llevas {errores} fallo(s).")

    print(f"\nHas perdido 😫. La palabra era: {palabra}")

````

````python
ahorcado()
````
