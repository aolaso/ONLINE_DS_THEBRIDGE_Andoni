## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
historial = []  # Nuestra lista para guardar lo que vamos haciendo

while True:
    print("\n--- MENÚ ---")
    print("Opciones: suma, resta, multi, div, potencia, historial, salir")
    
    # .lower() para que "SUMA" o "suma" funcionen igual
    opcion = input("Elige una operación: ").lower().strip()

    if opcion == "salir":
        print("Cerrando calculadora...")
        break
        
    if opcion == "historial":
        print("\n--- OPERACIONES GUARDADAS ---")
        if not historial:
            print("No hay nada guardado todavía.")
        else:
            for operacion in historial:
                print(operacion)
        continue # Volver al principio del bucle

    # Aquí empieza la parte de seguridad
    try:
        n1 = float(input("Primer número: "))
        n2 = float(input("Segundo número: "))
        
        # Guardaremos el resultado aquí
        resultado = None
        simbolo = ""

        if opcion == "suma":
            resultado = n1 + n2
            simbolo = "+"
        elif opcion == "resta":
            resultado = n1 - n2
            simbolo = "-"
        elif opcion == "multi":
            resultado = n1 * n2
            simbolo = "*"
        elif opcion == "div":
            resultado = n1 / n2 # Si n2 es 0, Python saltará directo al 'except'
            simbolo = "/"
        elif opcion == "potencia":
            resultado = n1 ** n2
            simbolo = "^"
        else:
            print("Esa operación no me la sé.")
            continue

        # Si el cálculo ha funcionado (resultado no es None)
        if resultado is not None:
            frase = f"{n1} {simbolo} {n2} = {resultado}"
            historial.append(frase) # Lo guardamos en el saco
            print(f"Resultado: {resultado}")

    # Si el usuario escribe letras en vez de números # Ayuda de IA
    except ValueError:
        print("Error: Solo se pueden meter números.")
    # Si el usuario intenta dividir por 0 # Ayuda de IA
    except ZeroDivisionError:
        print("Error: No se puede dividir entre cero, es imposible.")
```
