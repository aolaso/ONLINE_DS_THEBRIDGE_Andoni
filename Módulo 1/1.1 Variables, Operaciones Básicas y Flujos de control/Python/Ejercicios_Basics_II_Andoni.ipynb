## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math
````

````python
# Creo un mensaje de bienvenida al conversor (preferencia personal sobre donde mostrar el mensaje, en vez de en la celda de input) y le doy a la variante 'grados' el valor del input que ponga el usuario. 
print("Bievenidx al programa para convertir grados en radianes. Por favor, introduzca los grados:")
grados = input("Indique los grados aquí:")

# Intercambiamos la ',' que seguramente introduza el usuario en los grados por '.' para que el programa lo entienda. Añadimos un mensaje para ir guiando al usuario. 
grados = grados.replace(",",".")
print(f"Ha introducido {grados} grados")

# Convertimos el input del usuario, que lo tenemos en la variable 'grados' a 'float', ya que lo que se introdce en 'input' es siempre 'string'. 
grados = float(grados)

# Creamos el corazón del programa con la fórmula para la conversión de grados a radianes. Añadimos un mensaje mediante 'print f' para darle el resultado al usuario. 
radianes = grados * (math.pi/180)
print(f"El equivalente a {grados} grados en radianes es {radianes}")
````

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

````python
print("Vamos a calcular el área de un paralelogramo. Para ello, necesito que introduzcas la base y la altura en cm. Sigue los pasos a continuación:")

base = float(input("Introduce la base en cm").replace(',','.'))
altura = float(input("Introduce la altura en cm").replace(',','.'))

print(f"Has introducido los siguientes datos: {base} cm de base y {altura} cm de altura.")

area = round(base * altura, 3)

print(f'Con estos datos, el área del paralelogramo es de {area} cm²')
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
print("Introduce tu primer dato aquí:")
dato_1 = float(input("Introduce el primer dato").replace(",","."))
print("Has itroducido el siguiente valor:",dato_1)
print("\nIntroduce tu segundo dato aquí:")
dato_2 = float(input("Introduzca segundo valor").replace(",","."))
print("Has itroducido el siguiente valor:",dato_2)

suma = dato_1 + dato_2
print(f"\nLa suma de los datos es {suma}")

print("\nLa suma de los valores es igual a 10:",suma == 10)
print("\nLa suma de los valores es es superior a 10:", suma > 10)
print("\nLa suma de los valores es inferior a 10:", suma < 10)
````

## Ejercicio 8
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

Compruébalo después con python

1. False. Si tenemos un solo False en una cadena con AND, el resultado siemre va a ser False.
2. False. Si ejecutamos las sentencias en el orden que deberíamos por las paréntesis, nos da True and True. Si a eso le aplicamos NOT, es False.
3. False. La sentencia acaba en False con un AND, por lo que nos da igual lo que haya dentro de las paréntesis. 
4. False. Es True AND True AND False, por lo que como usamos AND en la sentencia y hay un False, es False el resultado. 

````python
print(True and True and False)
print(not ((True or False) and (True or False)))
print((False or False or False or True) and False)
print(not False and True and not True)
````
