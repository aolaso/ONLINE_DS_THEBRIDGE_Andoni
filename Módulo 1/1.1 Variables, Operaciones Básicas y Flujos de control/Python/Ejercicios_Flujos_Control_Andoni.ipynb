## Ejercicio 1
Declara una variable numérica que será una hora cualquiera del día. Implementa la siguiente casuística:

* Si es entre las 0 y las 8, print "Durmiendo"
* Si es entre las 9 y las 18, print "Trabajando"
* Si es entre las 19 y las 21, print "Clase"

```python
hora = 20

if hora >= 0 and hora <= 8:
    print("Está durmiendo")
elif hora >= 9 and hora <= 18:
    print("Está trabajando")
elif hora >= 19 and hora <= 21:
    print("Está en clase")
elif hora >= 22 and hora <= 24:
    print("Está de descanso")
else:
    print("Transporte o error")
```

## Ejercicio 9
En este ejercicio vamos a crear un pequeño juego. Se trata de intentar adivinar un numero del 1 al 5. Tenemos dos intentos para acertar. Pasos a seguir:

1. Ya viene implementado con la librería `random` cómo obtener un número aleatorio del 1 al 5
2. Tendrás que declarar en una variable el numero de vidas, y mediante un bucle while, comprobar que todavia quedan vidas.
3. Dentro del bucle, obtener el valor del usuario y comprobar si es ese el numero a adivinar. Si no, actualizar las vidas.
4. Si acertamos, salimos del bucle e imprimimos por pantalla "You win". Y si perdemos también salimos del bucle, pero en este caso imprimimos por pantalla "You lose".

TIP: te puede resultar útil usar la sentencia `else` cuando acabe el bucle `while`. Lo que haya dentro de ese `else` se ejecutará una vez acabe la ejecución del `while`. Lo podrás usar para cuando pierdas.

```python
import random

numero = random.randint(1, 5)
vidas = 2

while vidas > 0:
    prueba = int(input("Adivina, adivinanza. ¿En qué número del 1 al 5 estoy pensando?"))
    if prueba == numero:
        print("You win")
        break
    else:
        vidas = vidas - 1
else:
    print("You lose")
```

```python
import random

numero_maquina = random.randint(1, 5) # Lo llamo numero_maquina para que sea más claro
vidas = 2

while vidas > 0:
    # 1. Pedimos el número al usuario
    prueba = int(input("Adivina, adivinanza. ¿En qué número del 1 al 5 estoy pensando?: "))
    
    # 2. Comprobamos si ha acertado
    if prueba == numero_maquina:
        # Aquí printeamos el número que tú metiste y el mensaje de victoria
        print(f"Has elegido el {prueba}. ¡HAS ACERTADO!")
        print("You win")
        break
    else:
        # 3. Si falla, restamos vida y avisamos
        vidas = vidas - 1
        print(f"Has elegido el {prueba}, pero no es correcto. Te quedan {vidas} vidas.")

# Este 'else' pertenece al 'while'. Se ejecuta si las vidas llegan a 0 sin haber hecho 'break'
else:
    print("*" * 20)
    print("Se han acabado tus intentos.")
    # 4. Aquí revelamos qué número había pensado la máquina
    print(f"Yo estaba pensando en el número: {numero_maquina}")
    print("You lose")
```
