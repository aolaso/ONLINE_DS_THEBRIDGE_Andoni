## Ejercicio 4
Realiza un conversor de grados Celsius a Fahrenheit. El conversor tiene que recibir el input del usuario y devolver la conversión.

La conversión de grados sigue la siguiente fórmula: F = C × (9 / 5) + 32 

```python
print("Bienvenidx al conversor de grados Celsius a Fahrenheit. Por favor, introduce los grados Celsius que quieras convertir:")
dato_usuario = input()
grados_celsius = float(dato_usuario)

grados_fah = grados_celsius * (9/5) + 32

print(f"Has introducido {grados_celsius}ºC.\nEsto equivale a {grados_fah}ºF.")
```

## Ejercicio 5
Comenta  el Ejercicio 4. Pon un comentario multilinea y haz comentarios de cada linea de código

```python
''' 
Empezamos por generar el texto que va a aparecer "pintado" en la pantalla para dar la bienvenida al conversos de Celsius a Fahrenheit
'''
print("Bienvenidx al conversor de grados Celsius a Fahrenheit. Por favor, introduce los grados Celsius que quieras convertir:")

''' 
Y ahora asignamos una variable al input (dato que va a añadir el usuario - que son grados Celsius)
'''  
dato_usuario = input()

''' 
Después creamos una nueva variable para convertir en float (número con decimales) el dato que introduzca el usuario para los grados Celsius
'''
grados_celsius = float(dato_usuario)

''' 
Generamos la variable para grados Fahrenheit usando la fórmula que se nos ha proporcionado. Utilizamos la variable de 'grados_celsius' (que es un float) para la fórmula. 
'''
grados_fah = grados_celsius * (9/5) + 32

''' 
Pintamos el mensaje final utilizando f string para poder introducir las variables de 'grados_celsius' y 'grados_fah' entre corchetes.
'''
print(f"Has introducido {grados_celsius}ºC.\nEsto equivale a {grados_fah}ºF.")
```
