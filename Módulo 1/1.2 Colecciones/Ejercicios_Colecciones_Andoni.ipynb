## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
ejer_3_resultado = []

for i in ejer_3_1:
    for j in ejer_3_2:
        ejer_3_resultado.append(i + " " + j)

print(ejer_3_resultado)
````

## Ejercicio 4
Dada la siguiente lista, encuentra por el índice el valor 45 y sustitúyelo por el 0

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

TIP: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]

# Usando comprensión, creo que es la forma más sencilla de llegar al objetivo, sin eliminar elementos de la lista original. 
ejer_5_results = [x for x in ejer_5 if x != 3]
print(ejer_5_results)
````

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]

# He visto que se puede hacer así preguntando, aunque no se me hubiera ocurrido

for value in ejer_5.copy():
    if value == 3:
        ejer_5.remove(value)

print(ejer_5)

3 in ejer_5
````

## Ejercicio 11
Convierte la siguiente lista en un diccionario
> ```ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
````

````python
print(dict(ejer_11)) # de esta manera no funciona porque hacemos que se sobrerescriba, por lo que no nos vale 
````

````python
# Lista original de tuplas
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]

# Creamos el diccionario vacío
mi_diccionario = {}

# Recorremos la lista usando las dos variables (letra y numero)
for letra, numero in ejer_11:
    
    # Si la letra NO está en el diccionario, la creamos con una lista vacía
    # Esto evita que Python nos de un error al intentar añadir números
    if letra not in mi_diccionario:
        mi_diccionario[letra] = []
    
    # Ahora añadimos el número a la lista de esa letra
    # Así no borramos el anterior, sino que los vamos acumulando
    mi_diccionario[letra].append(numero)

# Mostramos el resultado final
print(mi_diccionario)
````

## Ejercicio 15
Multiplica todos los valores del diccionario
> ```ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

````python
ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}
print(ejer_15)

list(ejer_15.values()) #creamos una lista de los valores para poder usarlos 
````

````python
ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}

# Empezamos con 1 porque si fuera 0, cualquier multiplicación daría 0
resultado = 1

# Usamos .values() para ignorar las claves (1, 2, 3...) y quedarnos solo con los números (11, 22...)
for valor in ejer_15.values():
    # Multiplicamos el acumulado por el nuevo valor y lo guardamos encima
    resultado = resultado * valor

print(f"El resultado total de la multiplicación es: {resultado}")
````
