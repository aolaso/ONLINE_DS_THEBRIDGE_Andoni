## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
import math
grados = float(input("Especifica los grados que quieres convertir a radianes"))
radianes = grados*(math.pi/180)
print("Los grados que has introducido equivalen a ", radianes, "radianes")
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
base = input("Especifica la base del paralelogramo en metros:")
base_real = float (base.replace (",", "."))
altura = input("Especifica la altura del paralelogramo en metros:")
altura_real = float (altura.replace (",", "."))
area = base_real * altura_real
print("El area del paralelogramo es", area, "metros cuadrados")
```
